```python
import csv
import random

grocery_items = [
    "Bread", "Milk", "Eggs", "Butter", "Cheese",
    "Yogurt", "Cereal", "Pasta", "Rice", "Tomatoes",
    "Potatoes", "Onions", "Apples", "Bananas", "Oranges",
    "Chicken", "Beef", "Fish", "Spinach", "Carrots"
]

num_transactions = 50

transactions = []
for _ in range(num_transactions):
    num_items = random.randint(2, 15)
    items = random.sample(grocery_items, num_items)
    transactions.append(items)
    
csv_file_path = "./data/market_basket_data.csv"
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Transaction", "Items"])
    for i, transaction in enumerate(transactions, start=1):
        writer.writerow([f"Transaction {i}", ','.join(transaction)])

print(f"CSV file created: {csv_file_path}")
```

```python
C = dict()

def count_frequency(item = None):
    if not item:
        return " Error "
    if type(item) == int:
        item = list([item])
    count = 0
    for transaction in data:
        if set(item).issubset(set(data[transaction])):
            count+=1
    return (count if '%' not in control_s_input else round(count*100/len(data),2))
```

```python
cache_C = set()
def print_table(C, F, step=0):
    global cache_C
    temp = cache_C
    C_col1 = list(C.keys())
    C_col1 = list(set(C_col1) - set(cache_C))
    C_col2 = [C[i] for i in C_col1]
    F_col1 = list(F.keys())
    F_col2 = list(F.values())
    df1 = pd.DataFrame({"itemset": C_col1, "support": C_col2})
    df2 = pd.DataFrame({"itemset": F_col1, "support": F_col2})
    print(f"Step: {step+1}")
    print(f"C{step}:",df1,"\n\n",f"F{step}:", df2)
    print("\n")
    cache_C = C_col1
    return
```

```python
F = dict()
for i in range(0, max_combinations):
    if len(F.keys())>0:        
        C = F
        
    # resetting F
    F = dict()
    F_temp = dict()
    
    # making F{i}
    for itemset in key_list:
        item_frequency = count_frequency(itemset)
        C[itemset] = item_frequency
        if item_frequency >= minimum_support:
            F_temp[itemset] = item_frequency 
    
    
    
    if len(F_temp.keys())>0:
        F = F_temp
    else:
        break
        
    # elimination
    key_list = list(F.keys())
    
    # making combinations
    temp = []
    for a in range(0, len(key_list)):
        for b in range(a+1, len(key_list)):
            temp.append(tuple(set(key_list[a]).union(set(key_list[b]))))
     
    key_list = temp
    print_table(C, F, i)
    

C_temp = dict(filter(lambda x: True if x[1] >= minimum_support and len(x[0]) == max_combinations else False, [i for i in C.items()]))
if (len(C_temp)) > 0:
    C = C_temp
else:
    C = dict(filter(lambda x: True if x[1]>=minimum_support else False, [i for i in C.items()]))
del cache_C
```

```python
def func(transactions, ant,con):
    total_transactions = len(transactions)
    support1=sum(1 for transaction in transactions.values() if set(ant).issubset(transaction))/total_transactions
    support2=sum(1 for transaction in transactions.values() if set(con).issubset (transaction))/total_transactions
    support=sum(1 for transaction in transactions.values() if set(ant).issubset(transaction) and set(con).issubset(transaction))/total_transactions
    confidence=support/support1
    lift=support/(support1*support2)
    return (support,confidence,lift)
```

```python
from itertools import combinations
def get_all_combinations(input_set):
    all_combinations = []
    for r in range(1, len(input_set)+1):
        all_combinations.extend(combinations(input_set, r))
        return_set = set(all_combinations) - set(input_set)
        
    return set(all_combinations) - {tuple(input_set)}

get_all_combinations({1,2})
```
